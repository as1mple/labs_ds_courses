```python
from sympy import *
from sympy.plotting import plot
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, b = symbols('x,b')
```

```python
y = 0.5*x + b
```

```python
scatter = [(2,1), (1,3), (3,4)]
```

```python
MSE = lambda func, sct: sum([(el[1] - func.subs(x, el[0]))**2 for el in sct])/len(sct)
```

```python
mse_equation = MSE(y, scatter)
simplify(mse_equation)
```

```python
plot(mse_equation);
```

```python
diff_mse_equation = diff(mse_equation)
diff_mse_equation
```

```python
plot(diff_mse_equation);
```

```python
b_find  = float(solve(diff_mse_equation)[0])
b_find
```

```python
f = lambda x: eval(str(y.subs(b,b_find)))
plt.xlim(-1, 5)
plt.ylim(-1, 5)
x_space = np.linspace(plt.xlim()[0], plt.xlim()[1] , 100)
plt.scatter([x[0] for x in scatter], [y[1] for y in scatter], c="red")
plt.axhline(0, color = 'blue')
plt.axvline(0, color = 'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"MSE   b   = {MSE(func=y.subs(b,b_find), sct=scatter)}")
plt.plot(x_space, f(x_space), color = 'green');
```
